#Assignment 1

Try Linear Regression just using numpy (Without Tensorflow/Pytorch or other torch library). You can optionally use sklearn (if you want)

```python
# Import Numpy & PyTorch
import numpy as np
import torch
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43],
[91, 88, 64],
[87, 134, 58],
[102, 43, 37],
[69, 96, 70]], dtype='float32')

# Target (apples)
targets = np.array([[56],
[81],
[119],
[22],
[103]], dtype='float32')

inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
print(targets)
```

```python
# Weights and biases
w = torch.randn(1, 3, requires_grad=True)
b = torch.randn(1, requires_grad=True)
print(w)
print(b)
```

```python
def model(x):
    return torch.mm(x, w.t(), out=None) +b
```

```python
# Generate predictions
preds = model(inputs)
print("Prediction : ",preds)
print("Targets :",targets)
```

```python
#MSE loss function
# import torch.nn as nn
def mse(P,T):
  diff = 0
  i=0
  sum=0
  n = len(T)
  for k in range(0,n) :
    i += 1
    diff = T[k] - P[k]
    square = diff * diff
    sum += square
  mean =  sum / i
  return mean
  
output = mse(preds, targets)
print("Loss : ",output)
```

```python
# Compute gradients
print("gradients : ",output.backward())
```

```python
# Gradients for weights
print(w)
print(w.grad)
```

```python
# Gradients for biases
print(b)
print(b.grad)
```

```python
# Adjust weights & reset gradients
with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()

print(w)
print(b)
```

```python
new_preds = model(inputs)
new_loss = mse(new_preds, targets)

print("New Loss : ",new_loss)
```

```python
# Train for 100 epochs
for i in range(100):
    preds = model(inputs)
    loss = mse(preds, targets)
    loss.backward()
    with torch.no_grad():
        w -= w.grad * 1e-5
        b -= b.grad * 1e-5
        w.grad.zero_()
        b.grad.zero_()
```

```python
# Calculate loss
preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43],
[91, 88, 64],
[87, 134, 58],
[102, 43, 37],
[69, 96, 70]], dtype='float32')

# Target (apples)
targets = np.array([[56],
[81],
[119],
[22],
[103]], dtype='float32')

inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
print(targets)
```

```python
# Weights and biases
w = torch.randn(1, 3, requires_grad=True)
b = torch.randn(1, requires_grad=True)
print(w)
print(b)
```

```python
def model(x):
    return x @ w.t() + b
```

```python
# Generate predictions
preds = model(inputs)
print("Prediction : ",preds)
print("Targets :",targets)
```

```python
import torch.nn as nn
loss = nn.MSELoss()
output = loss(preds, targets)
output.backward()
print("Loss :",output)
```

```python
# Gradients for weights
print(w)
print(w.grad)
# Gradients for biases
print(b)
print(b.grad)
```

```python
# Train for 100 epochs
for i in range(100):
    preds = model(inputs)
    loss = mse(preds, targets)
    loss.backward()
    with torch.no_grad():
        w -= w.grad * 1e-5
        b -= b.grad * 1e-5
        w.grad.zero_()
        b.grad.zero_()
```

```python
# Calculate loss
preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```
